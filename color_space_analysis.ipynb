```python
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
import math
import glob
```

```python
# get the file paths of cropped crosses...

path = './color_analysis'
files = glob.glob(os.path.join(path, '*.jpeg'))

files
```

```python
# read images and convert them to both hsv and ycrcb color space...

images = [cv2.imread(f) for f in files]
images_hsv = [cv2.cvtColor(img, cv2.COLOR_BGR2HSV) 
              for img in images]

images_ycrcb = [cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb) 
                for img in images]
```

```python
def visualize_channel(img_list, num_channel, path):
    
    # visualize the given channel from the image list...
    figure = plt.figure(figsize=(5,5))
    for num_img, img in enumerate(img_list):
        figure.add_subplot(3, 3, num_img + 1)
        plt.imshow(img[:,:,num_channel], cmap=plt.cm.binary)
        plt.xticks([])
        plt.yticks([])
        plt.title('test{}.png: nail - {}'.format(math.ceil((num_img + 1)/2), 
                                                 (num_img % 2) + 1))

    plt.tight_layout()
    plt.savefig(path, bbox_inches="tight")
    plt.show()
```

```python
print ('Visualizing the red channel ...')
visualize_channel(images, 2, './color_analysis/results/red_c.png')
```

```python
print ('Visualizing the green channel ...')
visualize_channel(images, 1, './color_analysis/results/green_c.png')
```

```python
print ('Visualizing the blue channel ...')
visualize_channel(images, 0, './color_analysis/results/blue_c.png')
```

```python
print ('Visualizing the heu channel ...')
visualize_channel(images_hsv, 0, './color_analysis/results/heu_c.png')
```

```python
print ('Visualizing the saturation channel ...')
visualize_channel(images_hsv, 1, './color_analysis/results/sat_c.png')
```

```python
print ('Visualizing the brightness channel ...')
visualize_channel(images_hsv, 2, './color_analysis/results/brig_c.png')
```

```python
print ('Visualizing the luminance channel ...')
visualize_channel(images_ycrcb, 0, './color_analysis/results/y_c.png')
```

```python
print ('Visualizing the difference between red and luminance channel ...')
visualize_channel(images_ycrcb, 1, './color_analysis/results/cr_c.png')
```

```python
print ('Visualizing the difference between green and luminance channel ...')
visualize_channel(images_ycrcb, 2, './color_analysis/results/cb_c.png')
```
